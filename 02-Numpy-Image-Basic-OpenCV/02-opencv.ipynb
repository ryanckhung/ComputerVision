# OpenCV

```python
# type the following if import CV2 has problem
# !pip3 install opencv-python

import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('01-puppy.jpg')
img
```

```python
img_bgr = cv2.imread('01-puppy.jpg')
plt.imshow(img_bgr)
```

```python
# opencv read image in BGR format, conver it to RGB by
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('01-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

# Resize the Image

```python
img_rgb.shape
```

```python
img_resize =cv2.resize(img_rgb,(1300,275))
plt.imshow(img_resize)
```

# Flip the Image

```python
img_flip = cv2.flip(img_rgb,1)  # try 0, 1, -1
plt.imshow(img_flip)
```
