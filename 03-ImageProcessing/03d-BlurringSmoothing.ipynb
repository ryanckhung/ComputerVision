# Blurring and Smoothing help to reduce the noise (usually combine with edge detection)

### https://setosa.io/ev/image-kernels/

```python
import warnings
warnings.filterwarnings('ignore')

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    img = cv2.imread('./DATA/Lenna.png').astype(np.float32) / 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(5,5))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
display_img(i)
```

```python
img = cv2.imread('./DATA/Lenna.png').astype(np.float32)/255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

# Gamma Correction - controls the overall brightness of an image

```python
img = load_img()
gamma = 3/8
effected_image = np.power(img, gamma)
display_img(effected_image)
```

# Creating a kernel

```python
kernel = np.ones(shape=(5,5),dtype=np.float32)/15
```

```python
kernel
```

```python
dst = cv2.filter2D(img,-1,kernel)
display_img(dst)
```

```python
display_img(img)
```

# Using OpenCV filter

```python
# ksize means kernel size
blurred_img = cv2.blur(img,ksize=(5,5))
display_img(blurred_img)
```

```python
# (15,15) means kernel size
blurred_img = cv2.blur(img,(15,15))
display_img(blurred_img)
```

```python
# (5,5) is the kernel size
blurred_img = cv2.GaussianBlur(img,(5,5),10)
display_img(blurred_img)
```

```python
# 5 is the kernel size means 5x5 square kernel
median = cv2.medianBlur(img,5)
display_img(median)
```
