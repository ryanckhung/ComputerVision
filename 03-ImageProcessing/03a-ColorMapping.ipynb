# Color Mapping

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('01-puppy.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```
