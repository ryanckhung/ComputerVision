# Thresholding - Convert Color Image to Grayscale -> value over threshold set it as max value; below threshold set it as min value

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./DATA/rainbow.jpg')
```

```python
plt.imshow(img)
```

```python
# 0 parameter means read it in black and white
img = cv2.imread('./DATA/rainbow.jpg',0)
plt.imshow(img,cmap='gray')
```

```python
# The following example: 127 is threshold; 255 is the maxvalue
# there are many type of threshold method; like cv2.THRESH_BINARY, cv2.THRESH_BINARY_INV
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(thresh1,cmap='gray')
```

# Adaptive Threshold

```python
img = cv2.imread("./DATA/crossword.jpg",0)

th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,15,15) # Play around with these last 2 numbers

plt.figure(figsize=(15,15))
plt.imshow(th2, cmap='gray')
```
